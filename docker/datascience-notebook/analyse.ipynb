```python
import pandas as pd
```

```python
with open('./work/data/res_data_5_50_many', 'r') as file:
    # Read the file and split by newline
    lines = file.read().split('\n')
```

```python
many = pd.Series([float(l) for l in lines if l] )
```

```python
many.describe()
```

```python
many.quantile(q=0.95)
```

```python
many_plot = many[(many < 300)].plot()
```
